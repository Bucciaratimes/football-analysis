```python
def plotScatterMap(df,playerId,ax,time=None):
    pdf = df[df['playerId']==playerId]    
    if time is not None:
        pdf = pdf[pdf["minute"]<time]
    x = pdf['x']
    y = pdf['y']
    endX = pdf['endX']
    endY = pdf['endY']
    ax.scatter(y,x,color="#dc2f02",s=100)  
    ax.scatter(y,x,color="#ffba08",s=150,alpha=.3)   
    
    cmap = mpl.colors.LinearSegmentedColormap.from_list('cmap', ["#13B9D6","#e76f51","#D61327"])
    color = cmap(pdf["xT"]/pdf["xT"].max())
    
    ax.scatter(y,x,color=color,s=100)  
    ax.scatter(y,x,color=color,s=350,alpha=.3)  
    
    for index, row in pdf.iterrows():
        if row["dist"] > 36.57:
            if row["outcomeType.displayName"] == "Successful":
                ax.annotate("",
                       xy=(row['endY'], row['endX']),xytext=(row["y"],row["x"]),
                       arrowprops={'arrowstyle':"-|>",
                                   'fc':'#D61327',
                                   'ec':'#dc2f02',
                                    "connectionstyle":"angle3, angleA = 0, angleB = 95"
                                    #"connectionstyle":"arc, angleA = 90, angleB = 0"
                                  })
            else:
                ax.annotate("",
                       xy=(row['endY'], row['endX']),xytext=(row["y"],row["x"]),
                       arrowprops={'arrowstyle':"-|>",
                                   'fc':"#7400b8",
                                   'ec':"#7400b8",
                                    "connectionstyle":"angle3, angleA = 0, angleB = 95"
                                    #"connectionstyle":"arc, angleA = 90, angleB = 0"
                                  })
        else:
            if row["outcomeType.displayName"] == "Successful":
                ax.annotate("",
                               xy=(row['endY'], row['endX']),xytext=(row["y"],row["x"]),
                               arrowprops={'arrowstyle':"-|>,head_width=.7,head_length=.9",
                                           'fc':'#76c893',
                                           'ec':'#76c893',
                                      },
                          )
            else:
                ax.annotate("",
                               xy=(row['endY'], row['endX']),xytext=(row["y"],row["x"]),
                               arrowprops={'arrowstyle':"->",
                                           'fc':"#7400b8",
                                           'ec':"#7400b8",
                                      },
                          )
    return None

def passSonerMap(df,playerId,ax):
    pdf = df[df['playerId']==playerId]
    
    colors = cmap(pdf['count']/pdf['count'].max())
    multiplier = 2*np.pi/24
    bars = ax.bar(pdf['angle_bin']*multiplier, 
                   pdf['avg_length'], 
                   width=0.2, 
                   bottom=0, 
                   alpha=0.9, 
                   color=colors,
                   zorder=3)
    
    ax.set_xticklabels([])
    ax.set_yticks([])
    # ax_sub.grid(True, alpha=.5)
    ax.grid(False)
    # ax_sub.spines['polar'].set_visible(True)
    ax.spines['polar'].set_visible(False)

    # ax_sub.spines['polar'].set_color(main_color)
    ax.patch.set_alpha(0)
#     return axes
    return ax   

def plotHeatMap(df,playerId,ax):
    pdf = df[df['playerId']==playerId]
    cmap = mpl.colors.LinearSegmentedColormap.from_list('cmap', ["#13B9D6","#D61327"])
    
    kde = sns.kdeplot(
        pdf['y'],
        pdf['x'],
        shade=True,
        shade_lowest=False,
        alpha=.9,
        n_lavels=10,
        cmap=cmap,
        ax=ax) 
#     pos.invert_xaxis()
    return None
```

```python
#Bin Data for xT

# We use a pandas method called .cut --> https://pandas.pydata.org/docs/reference/api/pandas.cut.html
# We say labels=False because we want the integer locator of the bin
df['x_bin'] = pd.cut(df['x'], bins=xT_cols, labels=False)
df['y_bin'] = pd.cut(df['y'], bins=xT_rows, labels=False)
df['endX_bin'] = pd.cut(df['endX'], bins=xT_cols, labels=False)
df['endY_bin'] = pd.cut(df['endY'], bins=xT_rows, labels=False)
df['start_zone_value'] = df[['x_bin', 'y_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
df['end_zone_value'] = df[['endX_bin', 'endY_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
```

```python
# now we find the difference between the start and end to get the xT value
df['xT'] = df['end_zone_value'] - df['start_zone_value']
```
