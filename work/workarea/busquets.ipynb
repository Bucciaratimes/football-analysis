```python
def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i + 1]) for i in range(len(nums_list))]


def plotShotMap(url):

    mpl.rcParams['xtick.color'] = text_color
    mpl.rcParams['ytick.color'] = text_color

    res = requests.get(url)
    soup = BeautifulSoup(res.content, "lxml")
    scripts = soup.find_all('script')
    strings = scripts[1].string
    index_start = strings.index("('") + 2
    index_end = strings.index("')")
    json_data = strings[index_start:index_end]
    json_data = json_data.encode('utf8').decode('unicode_escape')
    data = json.loads(json_data)

    x_under = []
    y_under = []
    min_under = []
    result_under = []
    xG_under = []
    player_under = []
    team_under = []
    home_data = data['h']
    away_data = data['a']
    names = ['x', 'y', 'minute', 'team', 'xG', 'player', 'result']

    for index in range(len(home_data)):
        for key in home_data[index]:
            if key == 'X':
                x_under.append(home_data[index][key])
            if key == 'Y':
                y_under.append(home_data[index][key])
            if key == 'minute':
                min_under.append(home_data[index][key])
            if key == 'h_team':
                team_under.append(home_data[index][key])
            if key == 'xG':
                xG_under.append(home_data[index][key])
            if key == 'player':
                player_under.append(home_data[index][key])
            if key == 'result':
                result_under.append(home_data[index][key])

    for index in range(len(away_data)):
        for key in away_data[index]:
            if key == 'X':
                x_under.append(1 - float(away_data[index][key]))
            if key == 'Y':
                y_under.append(1 - float(away_data[index][key]))
            if key == 'minute':
                min_under.append(away_data[index][key])
            if key == 'a_team':
                team_under.append(away_data[index][key])
            if key == 'xG':
                xG_under.append(away_data[index][key])
            if key == 'player':
                player_under.append(away_data[index][key])
            if key == 'result':
                result_under.append(away_data[index][key])

    df2 = pd.DataFrame([x_under,
                        y_under,
                        min_under,
                        team_under,
                        xG_under,
                        player_under,
                        result_under],
                       index=names)
    df2 = df2.T

    df2['x'] = pd.to_numeric(df2['x'])
    df2['y'] = pd.to_numeric(df2['y'])
    df2['xG'] = pd.to_numeric(df2['xG'])
    df2['minute'] = pd.to_numeric(df2['minute'])

    df2['x'] = df2['x'] * 100
    df2['y'] = df2['y'] * 100

    df2['xx'] = df2['y']
    df2['yy'] = df2['x']

    df2['x'] = df2['xx']
    df2['y'] = df2['yy']

    df2['x'] = df2['x'] * .8
    df2['y'] = df2['y'] * 1.2

    home_team = df2['team'].iloc[0]
    away_team = df2['team'].iloc[-1]

    h_xG = [0]
    a_xG = [0]
    h_min = [0]
    a_min = [0]

    for x in range(len(df2['xG'])):
        if df2['team'][x] == home_team:
            h_xG.append(df2['xG'][x])
            h_min.append(df2['minute'][x])
        if df2['team'][x] == away_team:
            a_xG.append(df2['xG'][x])
            a_min.append(df2['minute'][x])

    home_cum = nums_cumulative_sum(h_xG)
    away_cum = nums_cumulative_sum(a_xG)
    home_last = round(home_cum[-1], 2)
    away_last = round(away_cum[-1], 2)
    home_score = 0
    away_score = 0

    for x in range(len(df2['x'])):
        if df2['team'][x] == home_team:
            if df2['result'][x] == 'Goal':
                axes.scatter(
                    df2['y'][x],
                    df2['x'][x],
                    color=scatter_color,
                    s=df2['xG'][x] * 1000,
                    edgecolor='gold',
                    hatch=5 * '/',
                    linewidth=2,
                    alpha=1)
                home_score += 1
            else:
                axes.scatter(
                    df2['y'][x],
                    df2['x'][x],
                    color='#b6b9ea',
                    s=df2['xG'][x] * 1000,
                    edgecolor=scatter_color,
                    facecolor=background_color,
                    hatch=5 * '/',
                    marker='o',
                    alpha=.8)

        if df2['team'][x] == away_team:
            if df2['result'][x] == 'Goal':
                axes.scatter(
                    df2['y'][x],
                    df2['x'][x],
                    color=scatter_color,
                    s=df2['xG'][x] * 1000,
                    edgecolor='gold',
                    hatch=5 * '/',
                    linewidth=2,
                    alpha=1)
                away_score += 1
            else:
                axes.scatter(
                    df2['y'][x],
                    df2['x'][x],
                    color='#b6b9ea',
                    s=df2['xG'][x] * 1000,
                    edgecolor=scatter_color,
                    facecolor=background_color,
                    hatch=5 * '/',
                    marker='o',
                    alpha=.8)

    axes.annotate(
        f'{home_team}', xy=(
            80, 8), xytext=(
            73, 8), color='#b6b9ea', fontsize=38, alpha=0.2)
    axes.annotate(
        f'{away_team}', xy=(
            5, 70), xytext=(
            4, 70), color='#b6b9ea', fontsize=38, alpha=0.2)

    htext.fig_text(
        0.355,
        0.49,
        s=f'<{home_team}> {home_score} - {away_score} <{away_team}>',
        fontsize=24,
        color=text_color,
        highlight_colors=[
            home_color,
            away_color])
```
