## Play Time

```python
df = pd.read_csv('/work/assets/fbref/report/mac_summary.csv')
df_time = df[['Min','Date']]
df_time = df_time.dropna()
for idx, row in df_time.iterrows():
    if len(row['Min']) > 2:
        df_time = df_time.drop(index=idx,axis=0)

df_time.sort_values(by='Date')

# y = [0]
# for i in df_time['Min']:
#     y.append(y[-1]+int(i))

y = []
for i in df_time['Min']:
    y.append(int(i))
    
_y = []
for i,val in enumerate(y):
    ys = np.mean(y[i:i+3])
    _y.append(ys)
    
    
_y = _y[:-1]
```

```python
df = pd.read_csv('/work/assets/fbref/report/lallana_summary.csv')
df_time = df[['Min','Date']]
df_time = df_time.dropna()
for idx, row in df_time.iterrows():
    if len(row['Min']) > 2:
        df_time = df_time.drop(index=idx,axis=0)

df_time.sort_values(by='Date')

y2 = []
for i in df_time['Min']:
    y2.append(int(i))

_y2 = []
for i,val in enumerate(y2):
    y = np.mean(y2[i:i+3])
    _y2.append(y)

# y2 = [np.mean(y2[0:3]), np.mean(y2[3:6]), np.mean(y2[6:10]), 
#      np.mean(y2[10:13]), np.mean(y2[13:15]), np.mean(y2[15:22]),np.mean(y2[22:])]


_y2 = _y2[:-3]
```

```python
df = pd.read_csv('/work/assets/fbref/report/trossard_summary.csv')
df_time = df[['Min','Date']]
df_time = df_time.dropna()
for idx, row in df_time.iterrows():
    if len(row['Min']) > 2:
        df_time = df_time.drop(index=idx,axis=0)

df_time.sort_values(by='Date')

y3 = []
for i in df_time['Min']:
    y3.append(int(i))

_y3 = []
for i,val in enumerate(y3):
    y = np.mean(y3[i:i+6])
    i+=4
    if i == len(y3)-2:
        break
    _y3.append(y)


_y3 = _y3[:-3]
```
