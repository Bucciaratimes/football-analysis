```python
import numpy as np
import pandas as pd 
import requests 
from bs4 import BeautifulSoup

import json

import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.colors import Normalize
from matplotlib import cm
          
    
base_url = 'https://understat.com/league'
leagues = [
    'La_liga', 'EPL', 'Bundesliga', 
    'Serie_A', 'Ligue_1', 'RFPL'
]
seasons = [
    '2014', '2015', '2016',
    '2017', '2018', '2019',
    '2020',
          ]

url = base_url + '/' + leagues[3] + '/' + seasons[-1]
html = requests.get(url)
soup = BeautifulSoup(html.content, 'lxml')
scripts = soup.find_all('script')

string_with_json_obj = scripts[3].string
index_start = string_with_json_obj.index("('") + 2
index_end = string_with_json_obj.index("')")
json_data = string_with_json_obj[index_start:index_end]
json_data = json_data.encode('utf8').decode('unicode_escape')
data = json.loads(json_data)
```

```python
df = {
    'id' : [],
    'player_name':[],
    'games' : [],
    'time' : [],
    'goals' : [],
    'xG' : [],
    'assists' : [],
    'xA': [],
    'shots' : [],
    'key_passes' : [],
    'position' : [],
    'team_title' : [],
    'npg': [],
    'npxG' : [],
    'xGChain': [],
    'xGBuildup' : []
}

for dat in data:
    for key in df.keys():
        df[key].append(dat[key])
        
```

```python
df = pd.DataFrame(df)
df.to_csv(f"/work/assets/understats/serie2021.csv")
```
