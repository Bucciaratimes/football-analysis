```python
# base_url = base_url + category + '?' +'offset='

columns = ['Id', 'Name', 'Age', 'Nationality', 'Club', "Height", "Weight", 'Value', 'Wage', 'Overall', 'Potential']
df = pd.DataFrame(columns = columns)
for i in range(0,2160,60):
    url = base_url + category + '?type=all&hih=220&wih=130&offset=' + str(i)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    table = soup.find('tbody')
    rows = table.find_all('tr')

    for row in rows:
        td = row.find_all('td')
        _id = td[0].find('img').attrs.get('id')
        name = td[1].find("a").attrs.get("aria-label")
        nation = td[1].find('img').attrs.get('title')
        age = td[2].text
        overall = td[3].text.strip()
        potential = td[4].text.strip()
        club = td[5].find('a').text
        height = td[6].text[:-2]
        weight = td[7].text[:-2]
        value = td[8].text.strip()
        wage = td[9].text.strip()

        
        _df = pd.DataFrame([[_id, name, age, nation, club, height, weight, value, wage, overall, potential]],columns=columns)
#         player_data.columns = columns
        df = df.append(_df, ignore_index=True)
    print("done for "+str(offset),end="\r")


df = df.drop_duplicates()
df = df.reset_index(drop=True)
```
