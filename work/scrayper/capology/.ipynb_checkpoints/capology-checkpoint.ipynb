```python
sns.set(style="darkgrid")
season = 2021
league = "La Liga"
df1 = pd.read_csv(f"/work/assets/capology/salary/{league}/{season}/{season}_La Liga.csv")
col = "Annual Gross" 
df1[col]= df1[col].str.replace(',', '').str.lstrip('€').str.lstrip().astype('float') / 1000000
sturges = lambda n: math.ceil(math.log2(n*2))
struges_bin = sturges(len(df1[col]))
print(struges_bin)

season = 2022
league = "La Liga"
df2 = pd.read_csv(f"/work/assets/capology/salary/{league}/{season}/{season}_La Liga.csv")
col = "Annual Gross" 
df2[col]= df2[col].str.replace(',', '').str.lstrip('€').str.lstrip().astype('float') / 1000000
sturges = lambda n: math.ceil(math.log2(n*2))
struges_bin = sturges(len(df2[col]))
print(struges_bin)

season = 2023
league = "La Liga"
df3 = pd.read_csv(f"/work/assets/capology/salary/{league}/{season}/{season}_La Liga.csv")
col = "Annual Gross" 
df3[col]= df3[col].str.replace(',', '').str.lstrip('€').str.lstrip().astype('float') / 1000000
sturges = lambda n: math.ceil(math.log2(n*2))
struges_bin = sturges(len(df3[col]))
print(struges_bin)
```

```python
spaceMono_bold = 'https://github.com/googlefonts/spacemono/blob/main/fonts/SpaceMono-BoldItalic.ttf?raw=true'

from mplsoccer import FontManager
from matplotlib.colors import LinearSegmentedColormap

monoBFont = FontManager(spaceMono_bold)
```

```python
team = "Barcelona"
```

```python
for param in ['text.color', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '#fdc526'  

fig,axes = plt.subplots(3,1,figsize=(18,27))
fig_color = "#171733"
fig.set_facecolor(fig_color)

dfs = [df1[df1["Club"]==team],df2[df2["Club"]==team],df3[df3["Club"]==team]]
axes = axes.flatten()

sns.barplot(y="Player", x=col, orient="h", data=df1[df1["Club"]==team], palette=sns.color_palette("cool_r", 30), ax=axes[0])
sns.barplot(y="Player", x=col, orient="h", data=df2[df2["Club"]==team], palette=sns.color_palette("spring_r", 35),ax=axes[1])
sns.barplot(y="Player", x=col, orient="h", data=df3[df3["Club"]==team], palette=sns.color_palette("winter_r", 31),ax=axes[2])

titles = ["20/21", "21/22", "22/23"]
for i in range(len(axes)):
    names = dfs[i]["Player"].unique()
    print(len(names))
    axes[i].set_xlim((0,70))
    axes[i].set_facecolor(fig_color)
    axes[i].xaxis.label.set_color('white')
    axes[i].yaxis.label.set_color('white')  
    axes[i].xaxis.label.set_fontsize(14)  
    axes[i].set_xlabel("Annual Gross",fontsize=18,fontweight='bold',color="#fdc526",fontproperties=monoBFont.prop)
    axes[i].set_ylabel("",fontweight='bold')   
    axes[i].set_yticklabels(names,fontsize=15,fontweight="bold",fontproperties=monoBFont.prop)
    axes[i].set_title(titles[i], fontweight='bold', fontsize=17, color="white")
    for label in (axes[i].get_xticklabels()):
        label.set_fontsize(16)
    
fig_text(s=f"<{team} Player's yearly Gross Salary>",
         x=axes[0].get_position().x0+.01,y=axes[0].get_position().y1+0.035,
         color="#ffffff",
         highlight_colors=["#fdc526"],
         highlight_weights=["bold"],
         highlight_styles=["italic"],
         fontsize=32,
         fontproperties=monoBFont.prop,fig=fig)

fig_text(s=f"<Data via capology | Created by @Bucciaratimes>",
         x=axes[0].get_position().x0+0.45,y=axes[0].get_position().y1+0.02,
         color="#f8f8f8",
         highlight_colors=["#fdc526"],
         highlight_weights=["bold"],
         highlight_styles=["italic"],
         fontsize=15,
         fontproperties=monoBFont.prop,fig=fig)
# plt.savefig(f'/work/output/{team}_salary.png', dpi=300, bbox_inches="tight")
```

```python
data = {"Club":["Barcelona","Barcelona","Barcelona",
                "Real Madrid","Real Madrid","Real Madrid",
                "Atletico Madrid","Atletico Madrid","Atletico Madrid"],
        "Season":["20/21","21/22","22/23","20/21","21/22","22/23","20/21","21/22","22/23"],
        "Value":[df1[df1["Club"]==team][col].sum(),
                 df2[df2["Club"]==team][col].sum(),
                 df3[df3["Club"]==team][col].sum(),
                 df1[df1["Club"]=="Real Madrid"][col].sum(),
                 df2[df2["Club"]=="Real Madrid"][col].sum(),
                 df3[df3["Club"]=="Real Madrid"][col].sum(),
                 df1[df1["Club"]=="Atletico Madrid"][col].sum(),
                 df2[df2["Club"]=="Atletico Madrid"][col].sum(),
                 df3[df3["Club"]=="Atletico Madrid"][col].sum()]}
df = pd.DataFrame(data=data)

sns.catplot(
    data=df, kind="bar",
    x="Season", y="Value", hue="Club",
    palette=sns.color_palette("cool_r",3),    
)

fig = plt.gcf()
fig_color = "#171733"
fig.set_facecolor(fig_color)
axes = plt.gca()
axes.set_facecolor(fig_color)
axes.xaxis.label.set_fontsize(14)  
axes.set_xlabel("Season",fontsize=15,fontweight='bold',color="#fdc526",fontproperties=monoBFont.prop)
axes.set_ylabel("Annual Gross",fontsize=15,fontweight='bold',color="#fdc526",fontproperties=monoBFont.prop)
axes.set_title("Total Salary", fontweight='bold', fontsize=18, color="#fdc526",fontproperties=monoBFont.prop)
# plt.savefig(f'/work/output/total_salary.png', dpi=300, bbox_inches="tight")
```
